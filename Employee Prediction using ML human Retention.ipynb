# Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# darker to lower color boxes shows the comparsion that for which factor someone left the company

def plot_corr(df,size=10): # correlation box
    corr=df.corr() # datafram.correlation
    fig,ax=plt.subplots(figsize=(size,size)) #figure size
    ax.legend()
    cax=ax.matshow(corr)
    fig.colorbar(cax)
    plt.xticks(range(len(corr.columns)), corr.columns, rotation='vertical')
    plt.yticks(range(len(corr.columns)), corr.columns)
    
plot_corr(main_df)
```

# KNN

```python
# Now Data Processing of kNN
```

```python
# Standard scalar
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Train the X_train into Standard Scalar and then calculate Standard deviation on X_test and X_train according to that data
sc=StandardScaler().fit(X_train)
X_train_std=sc.transform(X_train)
X_test_std=sc.transform(X_test)
```

```python
X_train_std
```

```python
X_test_std
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_std,y_train)
```

```python
prediction_knn=knn.predict(X_test_std) # predicted data of X_test
```

```python
# compare actual y_test data with predicted data, how much answers are matching
accuracy_knn=accuracy_score(y_test,prediction_knn)*100
```

```python
accuracy_knn
```

```python
prediction_knn
```

```python
y_test
```

```python
# Checking how much K value is best
k_range=range(1,26)
scores={}
scores_list=[]

for k in k_range:
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_std,y_train)
    prediction_knn=knn.predict(X_test_std)
    scores[k]=accuracy_score(y_test,prediction_knn)*100
    scores_list.append(accuracy_score(y_test,prediction_knn))
    
```

```python
scores 
```

```python
scores_list # why stored in list because it is used to plot the data 
```

```python
# this graph shows which value of k should be best
plt.plot(k_range,scores_list) 
```
